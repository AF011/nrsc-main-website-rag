```python
# PDF Files Loading and Partitioning - Extract images, tables, and chunk text

from unstructured.partition.pdf import partition_pdf
import time

all_elements = []

for pdf_path in pdf_files:

    file_start_time = time.time()

    elements = partition_pdf(
        filename=str(pdf_path),
        extract_images_in_pdf=False,
        strategy = "hi_res",
        #strategy = "fast",
        hi_res_model_name="yolox",
        infer_table_structure=True,
        chunking_strategy="by_title",
        max_characters=3000,
        new_after_n_chars=3800,
        combine_text_under_n_chars=200,
        #extract_image_block_output_dir=path,
        multipage_sections=True,  # TOC hierarchy
    )

    # Attach source metadata manually (VERY IMPORTANT)
    for el in elements:
        el.metadata.source = pdf_path.name

    all_elements.extend(elements)


    file_end_time = time.time()

    print(f"Processed: {pdf_path.name}")
    print(f"Elements extracted: {len(elements)}")
    print(f"Time taken: {file_end_time - file_start_time:.2f} seconds")
    print("-" * 50)

print(f"\nTotal elements extracted from all PDFs: {len(all_elements)}")
```

```python
def clean_metadata(md: dict) -> dict:
    clean = {}
    for k, v in md.items():
        # Simple types are allowed by Chroma
        if isinstance(v, (str, int, float, bool)) or v is None:
            clean[k] = v
        # Convert 1-element lists like ["eng"] -> "eng"
        elif isinstance(v, list) and len(v) == 1 and isinstance(v[0], (str, int, float, bool)):
            clean[k] = v[0]
        # Otherwise, drop or stringify (choose one)
        # else:
        #     clean[k] = str(v)
    return clean
```

```python
from langchain_core.documents import Document

min_len = 50  # or whatever threshold you like

docs = []
for el in all_elements:
    text = getattr(el, "text", None)
    if not text or len(text) < min_len:
        continue

    # Start from full Unstructured metadata
    md = el.metadata.to_dict()

    # Add/override anything you care about
    md["category"] = el.category
    # Optionally ensure a simple "source" field exists
    md.setdefault("source", el.metadata.source)

    docs.append(Document(page_content=text, metadata=md))
```

```python
clean_docs = [
    Document(page_content=d.page_content, metadata=clean_metadata(d.metadata))
    for d in docs
]
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1500,      # max characters per chunk
    chunk_overlap=200,    # small overlap for context
)

chunk_docs = splitter.split_documents(clean_docs)
print(len(docs), "original chunks")
print(len(chunk_docs), "after splitting")
```

```python
# LLM Integration

from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Initializing the Qwen2.5:1.5b model
llm = OllamaLLM(
    model="llama3.2:3b",
    temperature=0.0,
    top_p=1.0
)
```

```python
import pandas as pd

print("Extracting all chunks from Chroma...")

# Pull everything - ids come by default
result = vectorstore.get(include=["documents", "metadatas"])

chroma_ids = result["ids"]
documents  = result["documents"]
metadatas  = result["metadatas"]

print(f"Total chunks found: {len(chroma_ids)}")

# Build registry with our own clean chunk_ids
records = []
for i, (chroma_id, text, meta) in enumerate(zip(chroma_ids, documents, metadatas)):
    chunk_id = f"chunk_{str(i+1).zfill(4)}"
    records.append({
        "chunk_id"   : chunk_id,
        "chroma_id"  : chroma_id,        # keep for reference, just in case
        "text"       : text,
        "source_file": meta.get("filename", ""),
        "page_number": meta.get("page_number", ""),
        "category"   : meta.get("category", ""),
    })

registry_df = pd.DataFrame(records)

# Filter out garbage chunks (too short to generate meaningful questions)
before = len(registry_df)
registry_df = registry_df[registry_df["text"].str.len() >= 100].reset_index(drop=True)
after = len(registry_df)

print(f"Filtered out {before - after} short chunks. Remaining: {after}")

# Save
registry_df.to_csv("chunks_registry.csv", index=False)
print("Saved to chunks_registry.csv")

registry_df.head(3)
```

```python
import time
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Load registry
registry_df = pd.read_csv("chunks_registry.csv")

# Load existing progress if any (resume capability)
try:
    eval_df = pd.read_csv("eval_dataset.csv")
    done_chunk_ids = set(eval_df["chunk_id"].unique())
    print(f"Resuming — {len(done_chunk_ids)} chunks already processed.")
except FileNotFoundError:
    eval_df = pd.DataFrame()
    done_chunk_ids = set()
    print("Starting fresh.")

# Filter only unprocessed chunks
remaining = registry_df[~registry_df["chunk_id"].isin(done_chunk_ids)].reset_index(drop=True)
print(f"Chunks remaining to process: {len(remaining)}")

# Prompt
prompt = ChatPromptTemplate.from_template("""
You are an expert question generator for retrieval evaluation.

Given the passage below, generate exactly 2 specific factual questions that:
- Can only be answered using this passage
- Are clear and self-contained
- Do not reference "the passage" or "the document"
- Are different from each other

Return ONLY the 2 questions, one per line, no numbering, no extra text.

Passage:
{text}
""")

chain = prompt | llm | StrOutputParser()

# Generation loop with checkpointing
records = []
checkpoint_every = 50

for i, row in remaining.iterrows():
    try:
        response = chain.invoke({"text": row["text"][:1500]})  # cap at 1500 chars for speed
        
        lines = [q.strip() for q in response.strip().split("\n") if q.strip()]
        questions = lines[:2]  # take only first 2 no matter what
        
        if len(questions) < 2:
            print(f"[WARN] chunk {row['chunk_id']} returned only {len(questions)} question(s). Skipping.")
            continue
            
        for q in questions:
            records.append({
                "question"        : q,
                "chunk_id"        : row["chunk_id"],
                "source_file"     : row["source_file"],
                "page_number"     : row["page_number"],
                "ground_truth_text": row["text"]
            })
        
        # Checkpoint
        if (i + 1) % checkpoint_every == 0:
            checkpoint_df = pd.DataFrame(records)
            combined = pd.concat([eval_df, checkpoint_df], ignore_index=True)
            combined.to_csv("eval_dataset.csv", index=False)
            eval_df = combined
            records = []
            print(f"[{i+1}/{len(remaining)}] Checkpoint saved. Total so far: {len(eval_df)}")
            
    except Exception as e:
        print(f"[ERROR] chunk {row['chunk_id']}: {e}")
        continue

# Save final remaining records
if records:
    checkpoint_df = pd.DataFrame(records)
    combined = pd.concat([eval_df, checkpoint_df], ignore_index=True)
    combined.to_csv("eval_dataset.csv", index=False)
    print(f"Done. Total eval pairs saved: {len(combined)}")
else:
    print("Done. No new records to save.")
```

```python
# Loading eval dataset and initializing results store

import pandas as pd
import json
from tqdm import tqdm

# Load eval dataset
eval_df = pd.read_csv("eval_dataset.csv")
print(f"Total questions: {len(eval_df)}")

# Load existing eval results if resuming
try:
    results_df = pd.read_csv("eval_results.csv")
    done_questions = set(zip(results_df["question"], results_df["retriever"]))
    print(f"Resuming — {len(results_df)} results already saved.")
except FileNotFoundError:
    results_df = pd.DataFrame()
    done_questions = set()
    print("Starting fresh evaluation.")
```

```python
# Evaluation

records = []
checkpoint_every = 100

retrievers = {
    "similarity" : sim_retriever,
    "mmr"        : mmr_retriever,
    "threshold"  : threshold_retriever,
}

for i, row in tqdm(eval_df.iterrows(), total=len(eval_df)):
    question        = row["question"]
    ground_truth    = row["ground_truth_text"]
    chunk_id        = row["chunk_id"]
    
    for retriever_name, retriever in retrievers.items():
        
        # Skip if already done
        if (question, retriever_name) in done_questions:
            continue
        
        try:
            docs = retriever.invoke(question)
            
            # Extract page contents
            retrieved_texts = [doc.page_content for doc in docs]
            
            # Check if ground truth is in results
            hit        = 0
            rank       = None
            n_returned = len(retrieved_texts)
            
            for r, text in enumerate(retrieved_texts):
                if text.strip() == ground_truth.strip():
                    hit  = 1
                    rank = r + 1  # 1-indexed
                    break
            
            records.append({
                "question"      : question,
                "chunk_id"      : chunk_id,
                "retriever"     : retriever_name,
                "hit"           : hit,
                "rank"          : rank,       # None if not found
                "n_returned"    : n_returned, # important for threshold retriever
            })
            
        except Exception as e:
            print(f"[ERROR] retriever={retriever_name}, q={question[:50]}: {e}")
            continue
    
    # Checkpoint
    if (i + 1) % checkpoint_every == 0:
        checkpoint_df = pd.DataFrame(records)
        combined = pd.concat([results_df, checkpoint_df], ignore_index=True)
        combined.to_csv("eval_results.csv", index=False)
        results_df = combined
        records = []
        print(f"[{i+1}/{len(eval_df)}] Checkpoint saved. Total so far: {len(results_df)}")

# Save final
if records:
    checkpoint_df = pd.DataFrame(records)
    combined = pd.concat([results_df, checkpoint_df], ignore_index=True)
    combined.to_csv("eval_results.csv", index=False)
    print(f"Done. Total results saved: {len(combined)}")
else:
    print("Done.")
```

```python
# Evaluation Metrics

results_df = pd.read_csv("eval_results.csv")

metrics = {}

for retriever_name, group in results_df.groupby("retriever"):
    
    total       = len(group)
    hit_rate    = group["hit"].sum() / total
    
    # MRR — 1/rank if found, 0 if not
    mrr = group["rank"].apply(lambda r: 1/r if pd.notna(r) else 0).mean()
    
    # Precision@k — 1/n_returned if hit, else 0
    precision = group.apply(
        lambda r: 1/r["n_returned"] if r["hit"] == 1 and r["n_returned"] > 0 else 0, axis=1
    ).mean()
    
    # Coverage — % of questions where at least 1 result was returned
    coverage = (group["n_returned"] > 0).sum() / total
    
    metrics[retriever_name] = {
        "Hit Rate"   : round(hit_rate, 4),
        "MRR"        : round(mrr, 4),
        "Precision@k": round(precision, 4),
        "Coverage"   : round(coverage, 4),
        "Total Qs"   : total,
    }

metrics_df = pd.DataFrame(metrics).T
metrics_df.to_csv("eval_metrics.csv")
metrics_df
```
