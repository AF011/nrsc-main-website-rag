````python
# Imports
from unstructured.partition.pdf import partition_pdf
from langchain_core.documents import Document
from pathlib import Path
import time
import pandas as pd
````

````python
# PDF Files Path
pdf_dir = Path("../../data/pdf/nrsc-pdfs/")
all_pdfs = list(pdf_dir.glob("*.pdf"))

# Filter only relevant PDFs
pdf_files = [
    pdf for pdf in all_pdfs 
    if pdf.name.startswith("Chap") 
    or pdf.name.startswith("Bhuvan") 
    or pdf.name.endswith("Bhuvan.pdf") 
    or pdf.name.endswith("bhuvan.pdf")
]

print(f"Total PDFs found: {len(pdf_files)}")
for p in pdf_files:
    print(" -", p.name)
````

````python
# PDF Ingestion with Unstructured - hi_res strategy
all_elements = []

for pdf_path in pdf_files:
    file_start = time.time()
    
    elements = partition_pdf(
        filename=str(pdf_path),
        extract_images_in_pdf=False,
        strategy="hi_res",
        hi_res_model_name="yolox",
        infer_table_structure=True,
        chunking_strategy="by_title",   # ← unstructured chunks by section title
        max_characters=1500,            # ← max chunk size
        new_after_n_chars=1200,         # ← start new chunk after 1200 chars
        combine_text_under_n_chars=300, # ← combine tiny sections together
        multipage_sections=True,
    )
    
    for el in elements:
        el.metadata.source = pdf_path.name
    
    all_elements.extend(elements)
    
    print(f"---> {pdf_path.name} → {len(elements)} chunks ({time.time()-file_start:.1f}s)")

print(f"\nTotal elements: {len(all_elements)}")
````

- *strategy              → HOW to read the PDF (hi_res = smart layout detection)*
- *hi_res_model_name     → WHICH model detects layout (yolox)*
- *infer_table_structure → parse table rows/columns properly*
- *chunking_strategy     → HOW to group text into chunks (by section title)*
- *max_characters        → hard max size of each chunk (1500)*
- *new_after_n_chars     → soft limit, start looking for break point (1200)*
- *combine_text_under    → merge tiny sections so no garbage chunks (300)*
- *multipage_sections    → don't split a section at page boundary*

````python
# Cell 4 — Inspect chunk quality before filtering

# Check length distribution
lengths = [len(el.text) for el in all_elements if hasattr(el, 'text') and el.text]

import statistics
print(f"Total chunks      : {len(all_elements)}")
print(f"Min length        : {min(lengths)} chars")
print(f"Max length        : {max(lengths)} chars")
print(f"Average length    : {int(statistics.mean(lengths))} chars")
print(f"Median length     : {int(statistics.median(lengths))} chars")
print()

# Show chunks under 150 chars — these are garbage candidates
short = [el for el in all_elements if hasattr(el, 'text') and el.text and len(el.text) < 150]
print(f"Chunks under 150 chars: {len(short)}")
print()
print("Sample short chunks:")
for el in short[:10]:
    print(f"  [{el.category}] ({len(el.text)} chars) → {repr(el.text)}")
````

````python
# Cell 5 — Quality Filter

def is_quality_chunk(text):
    if not text:
        return False
    
    text = text.strip()
    
    # Too short — page numbers, headers, fragments
    if len(text) < 150:
        return False
    
    # Mostly URLs — no semantic value
    lines = text.split('\n')
    url_lines = sum(1 for l in lines if l.strip().startswith('http'))
    if url_lines / max(len(lines), 1) > 0.4:
        return False
    
    # Mostly numbers/symbols — garbled OCR or tables without context
    alpha_chars = sum(1 for c in text if c.isalpha())
    if alpha_chars / max(len(text), 1) < 0.4:
        return False
    
    return True

# Apply filter
before = len(all_elements)
clean_elements = [
    el for el in all_elements
    if hasattr(el, 'text') and is_quality_chunk(el.text)
]
after = len(clean_elements)

print(f"Before filtering : {before} chunks")
print(f"After filtering  : {after} chunks")
print(f"Removed          : {before - after} garbage chunks")
print()

# Show what got removed
removed = [
    el for el in all_elements
    if hasattr(el, 'text') and not is_quality_chunk(el.text)
]
print("Sample removed chunks:")
for el in removed[:10]:
    print(f"  ({len(el.text)} chars) → {repr(el.text[:100])}")
````

````python
# Cell 6 — Convert to LangChain Documents with clean metadata

from langchain_core.documents import Document

def clean_metadata(md: dict) -> dict:
    """Keep only Chroma-compatible types: str, int, float, bool"""
    clean = {}
    for k, v in md.items():
        if isinstance(v, (str, int, float, bool)) or v is None:
            clean[k] = v
        elif isinstance(v, list) and len(v) == 1 and isinstance(v[0], (str, int, float, bool)):
            clean[k] = v[0]
    return clean

# Convert
clean_docs = []
for el in clean_elements:
    md = el.metadata.to_dict()
    md["category"] = el.category
    md.setdefault("source", el.metadata.source)
    
    clean_docs.append(Document(
        page_content=el.text,
        metadata=clean_metadata(md)
    ))

print(f"Total clean documents ready: {len(clean_docs)}")
print()

# Quick sanity check — show one good chunk
sample = clean_docs[10]
print("Sample chunk:")
print(f"  Source   : {sample.metadata.get('source')}")
print(f"  Page     : {sample.metadata.get('page_number')}")
print(f"  Category : {sample.metadata.get('category')}")
print(f"  Length   : {len(sample.page_content)} chars")
print(f"  Preview  : {sample.page_content[:200]}")
````

````python
# Cell 6b — Deduplicate clean_docs

seen_texts = set()
deduped_docs = []

for doc in clean_docs:
    text = doc.page_content.strip()
    if text not in seen_texts:
        seen_texts.add(text)
        deduped_docs.append(doc)

print(f"Before dedup : {len(clean_docs)}")
print(f"After dedup  : {len(deduped_docs)}")
print(f"Removed      : {len(clean_docs) - len(deduped_docs)} duplicates")

# Replace clean_docs
clean_docs = deduped_docs
````

````python
# Cell 7 — Ingest into 2 Chroma vectorstores

from langchain_ollama import OllamaEmbeddings
from langchain_chroma import Chroma
import time

embedding_models = {
    "nomic" : OllamaEmbeddings(model="nomic-embed-text"),
    "bge"   : OllamaEmbeddings(model="bge-m3"),
}

vectorstores = {}

for name, embeddings in embedding_models.items():
    print(f"\n{'='*50}")
    print(f"Ingesting with {name}...")
    start = time.time()
    
    vs = Chroma.from_documents(
        documents=clean_docs,
        embedding=embeddings,
        persist_directory=f"./chroma_db-{name}",
        collection_name=f"nrsc_docs-{name}"
    )
    
    elapsed = time.time() - start
    mins = int(elapsed // 60)
    secs = int(elapsed % 60)
    
    vectorstores[name] = vs
    print(f"---> {name} done → {vs._collection.count()} chunks indexed in {mins}m {secs}s")

print(f"\n{'='*50}")
print("All 2 vectorstores ready!")
````

nomic-embed-text  → baseline, English focused, 768 dims, 2048 token limit

bge-m3            → multilingual, 1024 dims, 8192 token limit

Dimensions  = output size (how rich the embedding vector is)
              
              nomic=768, bge=1024
              
              MORE dimensions = richer representation

Context limit = how much input text the model can handle
              
              nomic=2048 tokens, bge=8192 tokens
              
              These are INDEPENDENT — not related to each other

````python
# Cell 12 — Find threshold properly with multiple queries

test_queries = [
    "What is Bhuvan and when did it start?",
    "What is remote sensing?",
    "How are floods monitored using satellites?",
    "What is NDVI and how is it calculated?",
    "What are the types of soil in India?",
    "How does cyclone formation happen?",
    "What is land use land cover mapping?",
    "How is groundwater detected using remote sensing?",
    "What is the role of ISRO in disaster management?",
    "What are the applications of SAR data?",
    "How is drought monitored using satellite data?",
    "What is the difference between active and passive sensors?",
    "What is photosynthesis in remote sensing context?",
    "How are landslides detected from satellite imagery?",
    "What is ocean color remote sensing?",
]

import numpy as np

for vs_name, vs in vectorstores.items():
    all_scores = []
    
    for query in test_queries:
        results = vs.similarity_search_with_score(query, k=10)
        scores = [score for _, score in results]
        all_scores.extend(scores)
    
    all_scores = sorted(all_scores)
    
    print(f"\n[{vs_name}] across {len(test_queries)} queries:")
    print(f"  Min score  : {min(all_scores):.4f}")
    print(f"  Max score  : {max(all_scores):.4f}")
    print(f"  Mean score : {np.mean(all_scores):.4f}")
    print(f"  25th pct   : {np.percentile(all_scores, 25):.4f}")
    print(f"  50th pct   : {np.percentile(all_scores, 50):.4f}")
    print(f"  75th pct   : {np.percentile(all_scores, 75):.4f}")
    print(f"  90th pct   : {np.percentile(all_scores, 90):.4f}")
````

Distance scores  → 0.28 to 1.09  (lower = better)

Relevance scores → 0.0  to 1.0   (higher = better)

nomic best relevance → 0.4252

bge   best relevance → 0.4510

This explains why threshold was returning 0 results.
We set threshold = 0.65 and 0.87 based on distance scores. But the threshold retriever uses relevance scores internally. Your best results only reach 0.45 relevance — so nothing ever passed 0.65 or 0.87.

````python
# Cell 12 — Correct threshold based on RELEVANCE scores

retrievers["nomic_threshold"] = vectorstores["nomic"].as_retriever(
    search_type="similarity_score_threshold",
    search_kwargs={"score_threshold": 0.38, "k": 20},
)

retrievers["bge_threshold"] = vectorstores["bge"].as_retriever(
    search_type="similarity_score_threshold",
    search_kwargs={"score_threshold": 0.40, "k": 20},
)

print("---> Threshold retrievers updated based on RELEVANCE scores")
print(f"nomic threshold → 0.38")
print(f"bge threshold   → 0.40")
````

````python
# Cell — Eval Dataset Generation (v2)

import pandas as pd
import time
from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# LLM
llm = OllamaLLM(
    model="llama3.2:3b",
    temperature=0.0,
)

# Prompt
prompt = ChatPromptTemplate.from_template("""
You are an expert in remote sensing, geospatial science, 
and earth observation technologies.

Given the passage below, generate exactly 2 specific factual 
questions that a domain expert or student would genuinely ask 
about this topic.

STRICT RULES:
- Questions must be answerable ONLY from this passage
- ONLY use information explicitly present in the passage
- Do NOT use outside knowledge or add extra information
- Do NOT ask about author names, publication years, journal names
- Do NOT ask about figure numbers or table numbers
- Do NOT reference "the passage" or "the text" in questions
- Do NOT hallucinate facts not present in the passage
- Focus ONLY on: concepts, definitions, processes,
  findings, numbers, applications, techniques
- Each question must be self-contained and clear
- Both questions must be different from each other

Return ONLY 2 questions, one per line.
No numbering. No extra text. No explanation.

Passage:
{text}
""")

chain = prompt | llm | StrOutputParser()

# Resume capability
try:
    eval_df = pd.read_csv("eval_v2/eval_dataset_v2.csv")
    done_ids = set(eval_df["chunk_id"].unique())
    print(f"Resuming — {len(done_ids)} chunks already done")
    print(f"Questions so far — {len(eval_df)}")
except FileNotFoundError:
    eval_df = pd.DataFrame()
    done_ids = set()
    print("Starting fresh")

# Build registry from clean_docs
registry = []
for i, doc in enumerate(clean_docs):
    chunk_id = f"chunk_{str(i+1).zfill(4)}"
    registry.append({
        "chunk_id"   : chunk_id,
        "text"       : doc.page_content,
        "source_file": doc.metadata.get("source", ""),
        "page_number": doc.metadata.get("page_number", ""),
    })

registry_df = pd.DataFrame(registry)
remaining = registry_df[
    ~registry_df["chunk_id"].isin(done_ids)
].reset_index(drop=True)

print(f"\nTotal chunks    : {len(registry_df)}")
print(f"Already done    : {len(done_ids)}")
print(f"Remaining       : {len(remaining)}")
print(f"Estimated time  : ~{len(remaining) * 4 // 60} mins")
print(f"Total questions : ~{len(remaining) * 2} questions")
print("\nStarting generation...")
````

````python
# Generation loop for questions

records = []
checkpoint_every = 50

for i, row in remaining.iterrows():
    try:
        response = chain.invoke({"text": row["text"][:1200]})
        
        lines = [
            q.strip() for q in response.strip().split("\n")
            if q.strip() and len(q.strip()) > 20
        ]
        questions = lines[:2]

        if len(questions) < 2:
            print(f"[WARN] {row['chunk_id']} → only {len(questions)} question(s), skipping")
            continue

        for q in questions:
            records.append({
                "question"          : q,
                "chunk_id"          : row["chunk_id"],
                "source_file"       : row["source_file"],
                "page_number"       : row["page_number"],
                "ground_truth_text" : row["text"],
            })

        # Checkpoint every 50 chunks
        if (i + 1) % checkpoint_every == 0:
            checkpoint_df = pd.DataFrame(records)
            combined = pd.concat([eval_df, checkpoint_df], ignore_index=True)
            combined.to_csv("eval_dataset_v2.csv", index=False)
            eval_df = combined
            records = []
            print(f"[{i+1}/{len(remaining)}] Saved — Total questions: {len(eval_df)}")

    except Exception as e:
        print(f"[ERROR] {row['chunk_id']}: {e}")
        continue

# Final save
if records:
    checkpoint_df = pd.DataFrame(records)
    combined = pd.concat([eval_df, checkpoint_df], ignore_index=True)
    combined.to_csv("eval_dataset_v2.csv", index=False)
    print(f"\n---> Done! Total questions saved: {len(combined)}")
else:
    print(f"\n---> Done! Total questions saved: {len(eval_df)}")
````

````python
# Remove questions from reference chunks
import pandas as pd

eval_df = pd.read_csv("eval_dataset_v2.csv")

# Filter out chunks that are clearly reference sections
reference_keywords = [
    "References\n", 
    "User Handbook",
    "et al.,",
    "Journal of",
    "Vol.",
    "pp.",
]

def is_reference_chunk(text):
    count = sum(1 for kw in reference_keywords if kw in text)
    return count >= 2

before = len(eval_df)
eval_df = eval_df[
    ~eval_df["ground_truth_text"].apply(is_reference_chunk)
].reset_index(drop=True)
after = len(eval_df)

print(f"Before : {before} questions")
print(f"After  : {after} questions")
print(f"Removed: {before - after} bibliography questions")

eval_df.to_csv("eval_dataset_v2.csv", index=False)
print("---> Saved clean eval dataset")
````

````python
# Also remove questions where ground truth is garbled OCR
def is_garbled(text):
    words = text.split()
    # Too many single character words = garbled
    single_chars = sum(1 for w in words if len(w) == 1)
    return single_chars / max(len(words), 1) > 0.3

before = len(eval_df)
eval_df = eval_df[
    ~eval_df["ground_truth_text"].apply(is_garbled)
].reset_index(drop=True)
after = len(eval_df)

print(f"Before : {before}")
print(f"After  : {after}")
print(f"Removed: {before - after} garbled chunks")

eval_df.to_csv("eval_dataset_v2.csv", index=False)
print("---> Saved final clean eval dataset")
````

You're going in circles! Let me recap the full chain:

```
bge-m3 via Ollama → NaN values → 500 error
         ↓
    Why NaN?
         ↓
bge-m3 is a multi-vector model (dense + sparse + colbert)
         ↓
Ollama uses llama.cpp which only understands simple BERT-style models
         ↓
GGUF conversion of bge-m3 loses/breaks the sparse & colbert heads
         ↓
llama.cpp tries to pool a partially broken model
         ↓
attention mask edge case → softmax(all -inf) → NaN
         ↓
Ollama can't JSON serialize NaN → 500 error
```

---

### The root cause in one line:
**bge-m3 is too complex for Ollama/llama.cpp to handle correctly** — it was never designed to run via llama.cpp, it was designed to run via HuggingFace's `FlagEmbedding` library.

---

### So your real options are:

1. **Use a simpler model in Ollama** that llama.cpp handles well — like `nomic-embed-text` or `mxbai-embed-large`

2. **Use bge-m3 via HuggingFace** — which requires solving the torch version issue

````python
# Metrics Computation

import pandas as pd

results_df = pd.read_csv("eval_results_v2.csv")

print(f"Total results: {len(results_df)}")
print(f"Retrievers   : {results_df['retriever'].unique().tolist()}")

metrics = {}

for retriever_name, group in results_df.groupby("retriever"):

    total    = len(group)
    hits     = group[group["hit"] == 1]

    # Hit Rate
    hit_rate = len(hits) / total

    # MRR — computed only on hits
    mrr = (1 / hits["rank"]).mean() if len(hits) > 0 else 0

    # Precision@k
    precision = group.apply(
        lambda r: 1 / r["n_returned"]
        if r["hit"] == 1 and r["n_returned"] > 0
        else 0, axis=1
    ).mean()

    # Coverage
    coverage = (group["n_returned"] > 0).sum() / total

    metrics[retriever_name] = {
        "Hit Rate"   : round(hit_rate, 4),
        "MRR"        : round(mrr, 4),
        "Precision@k": round(precision, 4),
        "Coverage"   : round(coverage, 4),
        "Total Qs"   : total,
    }

metrics_df = pd.DataFrame(metrics).T
metrics_df = metrics_df.sort_values("Hit Rate", ascending=False)
metrics_df.to_csv("eval_metrics_v2.csv")

print("\n=== FINAL RESULTS ===")
print(metrics_df.to_string())
````

````python
def rrf_fusion(query, retriever1, retriever2, k=60, top_n=10):
    results1 = retriever1.invoke(query)
    results2 = retriever2.invoke(query)

    scores = {}
    doc_map = {}

    for rank, doc in enumerate(results1):
        key = doc.page_content.strip()
        scores[key] = scores.get(key, 0) + 1 / (k + rank + 1)
        doc_map[key] = doc

    for rank, doc in enumerate(results2):
        key = doc.page_content.strip()
        scores[key] = scores.get(key, 0) + 1 / (k + rank + 1)
        doc_map[key] = doc

    sorted_keys = sorted(scores.keys(), key=lambda x: scores[x], reverse=True)
    return [doc_map[k] for k in sorted_keys[:top_n]]

# Test
test_query = "What is Bhuvan and when did it start?"
results = rrf_fusion(test_query, retrievers["bge_similarity"], retrievers["bm25"])

print(f"Fusion results: {len(results)}")
print(f"Top result: {results[0].page_content[:100]}")
````

````python
# Fusion RRF and Fusion Ensemble

eval_df = pd.read_csv("eval_dataset_v2.csv")
print(f"Total questions: {len(eval_df)}")
print("Running fusion evaluation...")
print("=" * 50)

records = []

for i, row in tqdm(eval_df.iterrows(), total=len(eval_df)):
    question     = row["question"]
    ground_truth = row["ground_truth_text"]
    chunk_id     = row["chunk_id"]

    # Test both fusion approaches
    fusion_configs = {
        "fusion_rrf"      : lambda q: rrf_fusion(q, retrievers["bge_similarity"], retrievers["bm25"]),
        "fusion_ensemble" : lambda q: ensemble_retriever.invoke(q),
    }

    for retriever_name, retriever_fn in fusion_configs.items():
        try:
            docs           = retriever_fn(question)
            retrieved_texts = [doc.page_content for doc in docs]

            hit        = 0
            rank       = None
            n_returned = len(retrieved_texts)

            for r, text in enumerate(retrieved_texts):
                if text.strip() == ground_truth.strip():
                    hit  = 1
                    rank = r + 1
                    break

            records.append({
                "question"   : question,
                "chunk_id"   : chunk_id,
                "retriever"  : retriever_name,
                "hit"        : hit,
                "rank"       : rank,
                "n_returned" : n_returned,
            })

        except Exception as e:
            print(f"\n[ERROR] {retriever_name} | {question[:40]}: {e}")
            records.append({
                "question"   : question,
                "chunk_id"   : chunk_id,
                "retriever"  : retriever_name,
                "hit"        : 0,
                "rank"       : None,
                "n_returned" : 0,
            })
            continue

    if (i + 1) % 100 == 0:
        print(f"[{i+1}/{len(eval_df)}] processed...")

# Save
fusion_df = pd.DataFrame(records)
fusion_df.to_csv("eval_results_fusion_v2.csv", index=False)
print(f"\nDone! Total results: {len(fusion_df)}")

# Compute metrics
print("\n=== FUSION RESULTS ===")
print(f"{'Retriever':<20} {'Hit Rate':>10} {'MRR':>10} {'Precision':>10} {'Coverage':>10}")

for retriever_name, group in fusion_df.groupby("retriever"):
    hits      = group[group["hit"] == 1]
    hit_rate  = len(hits) / len(group)
    mrr       = (1 / hits["rank"]).mean() if len(hits) > 0 else 0
    precision = group.apply(
        lambda r: 1 / r["n_returned"] if r["hit"] == 1 and r["n_returned"] > 0 else 0,
        axis=1
    ).mean()
    coverage  = (group["n_returned"] > 0).sum() / len(group)
    print(f"{retriever_name:<20} {hit_rate:>10.4f} {mrr:>10.4f} {precision:>10.4f} {coverage:>10.4f}")

print("\n=== COMPARISON WITH INDIVIDUAL RETRIEVERS ===")
print(f"{'Retriever':<20} {'Hit Rate':>10} {'MRR':>10}")
print(f"{'bge_similarity':<20} {0.9201:>10.4f} {0.8349:>10.4f}")
print(f"{'bm25':<20} {0.8801:>10.4f} {0.8816:>10.4f}")
````
